# QFT-Based Quantum Attention Demo

This notebook demonstrates the QFT-based quantum attention mechanism.

```python
import numpy as np
import matplotlib.pyplot as plt
from qft_attention import QFTAttentionLayer, MultiHeadQFTAttention
from qft_attention_advanced import QFTAttentionLayerAdvanced
from noise_simulation import NISQNoiseSimulator
from classical_comparison import AttentionComparator
print('Modules imported!')
```

## 1. Basic Quantum Attention

```python
n_tokens = 4
layer = QFTAttentionLayer(n_tokens=n_tokens, shots=1024)
embeddings = np.random.randn(n_tokens, n_tokens)
output, attention = layer(embeddings)
print(f'Output shape: {output.shape}')
print(f'Attention shape: {attention.shape}')
```

## 2. Visualizing Attention

```python
fig, ax = plt.subplots(figsize=(6, 5))
im = ax.imshow(attention, cmap='viridis')
ax.set_title('Quantum Attention Weights')
plt.colorbar(im, ax=ax)
plt.show()
```

## 3. Multi-Head Attention

```python
mha = MultiHeadQFTAttention(n_tokens=4, n_heads=2, shots=512)
output = mha(embeddings)
print(f'Multi-head output shape: {output.shape}')
```

## 4. Noise Simulation

```python
simulator = NISQNoiseSimulator(backend='ibm_brisbane')
circuit = layer._build_circuit()
fidelity = simulator.estimate_fidelity(circuit)
print(f'Estimated fidelity: {fidelity:.4f}')
```

## 5. Classical Comparison

```python
comparator = AttentionComparator()
result = comparator.compare(n_tokens=4, embed_dim=4)
print(f'KL Divergence: {result.kl_divergence:.4f}')
print(f'Correlation: {result.correlation:.4f}')
```

## Summary

This demo showed basic quantum attention, multi-head attention, noise simulation, and classical comparison.
